```python
from keras.datasets import boston_housing
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
((train_data, train_labels) , (test_data, test_label)) = boston_housing.load_data()
```

Do some preprocessing of data

```python
mean = np.mean(train_data,axis=0)
std = np.std(train_data, axis=0)

train_data -= mean
train_data /= std

test_data -=mean
test_data /= std

val_data = train_data[:100]
val_labels = train_labels[:100]

train_data_partial = train_data[100:]
train_labels_partial = train_labels[100:]
```

```python
def build_model():
    model = Sequential()
    model.add(Dense(64, activation='relu', input_shape=(train_data.shape[1], )))
    model.add(Dense(1))
    model.compile(optimizer='rmsprop', loss='mse', metrics=['mse', 'accuracy'])
    return model
```

```python
model = build_model()
model.fit(train_data_partial, train_labels_partial, validation_data=(val_data, val_labels), epochs=500, batch_size=100)
```

```python
model.evaluate(test_data, test_label,batch_size=40, verbose=1,
```
