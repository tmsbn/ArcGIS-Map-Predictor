```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten
from keras.layers.convolutional import Conv2D, MaxPooling2D
from keras.datasets import mnist
from keras.utils import to_categorical
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
from keras import layers
from keras import models
from keras import optimizers

import matplotlib.pyplot as plt
import os, shutil
from keras.models import load_model, model_from_json
from pathlib import Path
```

```python
original_dataset_dir = os.path.join(os.path.expanduser('~'),'Desktop', 'esri project', 'original data');
base_dir =  os.path.join(os.path.expanduser('~'), 'Desktop', 'esri project', 'data');
base_scripts_dir =  os.path.join(os.path.expanduser('~'), 'Desktop', 'esri project', 'scripts');

MODEL_FILE_NAME = 'building_model.h5'

BUILDING_DIR = 'buildings'
OCEAN_DIR = 'forest'

BUILDING_FILE_NAME = 'buildings'
OCEAN_FILE_NAME = 'forest'

train_dir = os.path.join(base_dir, 'train')
validation_dir = os.path.join(base_dir, 'validation')
test_dir = os.path.join(base_dir, 'test')

train_building_dir = os.path.join(train_dir, BUILDING_DIR)
train_ocean_dir = os.path.join(train_dir, OCEAN_DIR)

validation_building_dir = os.path.join(validation_dir, BUILDING_DIR)
validation_ocean_dir = os.path.join(validation_dir, OCEAN_DIR)

test_building_dir = os.path.join(test_dir, BUILDING_DIR)
test_ocean_dir = os.path.join(test_dir, OCEAN_DIR)

model_file_path = os.path.join(base_scripts_dir, MODEL_FILE_NAME);
print(model_file_path)

IMAGE_SIZE = 150

EPOCHS = 10
```

```python
def make_dir_helper(dir_path):
    if not os.path.exists(dir_path):
        os.mkdir(dir_path)
        return True
    return False
```

```python
make_dir_helper(base_dir)

# Make base directories
make_dir_helper(train_dir)
make_dir_helper(validation_dir)
make_dir_helper(test_dir)

# Make train sub directories
make_dir_helper(train_building_dir)
make_dir_helper(train_ocean_dir)

# Make validation sub directories
make_dir_helper(validation_ocean_dir)
make_dir_helper(validation_building_dir)

# Make test sub directories
make_dir_helper(test_building_dir)
make_dir_helper(test_ocean_dir)

ext = '.jpg'
```

```python
is_model_from_file = False
model_file = Path(model_file_path)
model = None
if model_file.is_file():
    print('found model!')
    model = load_model(model_file_path)
    is_model_from_file = True
else:
    print('No saved model')
```

```python
    # Copy buildings into the training dataset
    fnames = [(BUILDING_FILE_NAME + '{}' + ext).format(i) for i in range(60)]
    for fname in fnames:
        src_dir = os.path.join(original_dataset_dir, fname)
        dst_dir = os.path.join(train_building_dir, fname)
        shutil.copyfile(src_dir, dst_dir)


    # Copy building into the validation dataset
    fnames = [(BUILDING_FILE_NAME + '{}' + ext).format(i) for i in range(60, 100)]
    for fname in fnames:
        src_dir = os.path.join(original_dataset_dir, fname)
        dst_dir = os.path.join(validation_building_dir, fname)
        shutil.copyfile(src_dir, dst_dir)

    # Copy building into the test dataset
    fnames = [(BUILDING_FILE_NAME + '{}' + ext).format(i) for i in range(60, 100)]
    for fname in fnames:
        src_dir = os.path.join(original_dataset_dir, fname)
        dst_dir = os.path.join(test_building_dir, fname)
        shutil.copyfile(src_dir, dst_dir)

    # Copy ocean into the training dataset
    fnames = [(OCEAN_FILE_NAME + '{}' + ext).format(i) for i in range(60)]
    for fname in fnames:
        src_dir = os.path.join(original_dataset_dir, fname)
        dst_dir = os.path.join(train_ocean_dir, fname)
        shutil.copyfile(src_dir, dst_dir)

    # Copy ocean into the validation dataset
    fnames = [(OCEAN_FILE_NAME + '{}' + ext).format(i) for i in range(60, 100)]
    for fname in fnames:
        src_dir = os.path.join(original_dataset_dir, fname)
        dst_dir = os.path.join(validation_ocean_dir, fname)
        shutil.copyfile(src_dir, dst_dir)

    # Copy ocean into the test dataset
    fnames = [(OCEAN_FILE_NAME + '{}' + ext).format(i) for i in range(60, 100)]
    for fname in fnames:
        src_dir = os.path.join(original_dataset_dir, fname)
        dst_dir = os.path.join(test_ocean_dir, fname)
        shutil.copyfile(src_dir, dst_dir)
```

```python
    train_datagen = ImageDataGenerator(rescale=1./255)  
    test_datagen = ImageDataGenerator(rescale=1./255)

    train_generator = train_datagen.flow_from_directory(
            train_dir,  
            target_size=(IMAGE_SIZE, IMAGE_SIZE),  
            batch_size=20,
            class_mode='binary')  

    validation_generator = test_datagen.flow_from_directory(
            validation_dir,
            target_size=(IMAGE_SIZE, IMAGE_SIZE),
            batch_size=20,
            class_mode='binary')
```

```python
    for idx, (data_batch, labels_batch) in enumerate(train_generator):
        print('data batch:', data_batch.shape)
        print('labels batch:', labels_batch.shape)

        if idx > 5:
            break
```

```python
    model = models.Sequential()
    model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(IMAGE_SIZE, IMAGE_SIZE, 3)))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(64, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Conv2D(128, (3, 3), activation='relu'))
    model.add(layers.MaxPooling2D((2, 2)))
    model.add(layers.Flatten())
    model.add(layers.Dropout(0.5))
    model.add(layers.Dense(512, activation='relu'))
    model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',
            optimizer=optimizers.RMSprop(lr=1e-4),
             metrics=['acc'])
```

```python
model.summary()
```

```python
if not is_model_from_file:
    history = model.fit_generator(
          train_generator,
          steps_per_epoch=10,
          epochs=EPOCHS,
          validation_data=validation_generator,
          validation_steps=10)
```

```python
    model.save(MODEL_FILE_NAME) 
```

```python
    acc = history.history['acc']
    val_acc = history.history['val_acc']
    loss = history.history['loss']
    val_loss = history.history['val_loss']

    epochs = range(len(acc))

    plt.plot(epochs, acc, 'b+')
    plt.plot(epochs, val_acc, 'bo')
    plt.title('Training and validation accuracy')

    plt.figure()

    plt.plot(epochs, loss, 'b+')
    plt.plot(epochs, val_loss, 'bo')
    plt.title('Training and validation loss')

    plt.show()
```
